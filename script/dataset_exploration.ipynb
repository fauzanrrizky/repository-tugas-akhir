# 7M+ Dataset

```python
df = pd.read_csv('../data/companies_sorted.csv')
```

```python
countries = ['united states', 'germany', 'india', 'united kingdom', 'france', 'russia', 'canada', 'italy', 'brazil', 'australia', 'mexico', 'spain', 'indonesia', 'netherlands']
```

```python
country_counts = df['country'].value_counts().head(20)
# Plot a pie chart
plt.figure(figsize=(8, 6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Top 5 Country Count')
plt.show()
```
