```python
data_done = pd.read_csv('data/data_context/data_done.csv', on_bad_lines='skip')
data_with_context = pd.read_json('data/data_context/data_with_context.jsonl', lines=True)
```

```python
def update_context(data_done, data_with_context):
    for index, row in data_done.iterrows():
        # Jika c1_context adalah 'No Desc', cari nilainya di data_with_context berdasarkan c1 dan c2
        if row['c1_context'] == 'No Desc':
            matched_row = data_with_context[(data_with_context['c1'] == row['c1']) & (data_with_context['c2'] == row['c2'])]
            if not matched_row.empty:
                data_done.at[index, 'c1_context'] = matched_row['c1_context'].values[0]

        # Jika c2_context adalah 'No Desc', cari nilainya di data_with_context berdasarkan c1 dan c2
        if row['c2_context'] == 'No Desc':
            matched_row = data_with_context[(data_with_context['c1'] == row['c1']) & (data_with_context['c2'] == row['c2'])]
            if not matched_row.empty:
                data_done.at[index, 'c2_context'] = matched_row['c2_context'].values[0]

# Jalankan fungsi untuk mengupdate data_done
update_context(data_done, data_with_context)
```

```python
data_done.to_csv('data/data_context/temp.csv', index=False)
data_done.to_json('data/data_context/temp.jsonl', orient='records', lines=True)
```
